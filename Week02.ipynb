5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!

````python

import numpy as np
import random

cp5_img = img.copy()
height,width,_ = cp5_img.shape
image = np.copy(cp5_img)

for i in range(5):
    x1 = random.randint(0, width - 50)
    y1 = random.randint(0, height - 50)
    x2 = x1 + random.randint(20, 100)
    y2 = y1 + random.randint(20, 100)
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    cv.rectangle(image, (x1, y1), (x2, y2), color, -1)

cv2_imshow( image)
````
